# 페이지	제목	URL 수집하여 txt파일 저장

```python
import requests
from lxml import html
import time


keyword = 'pann'


# 파일 만들기
output_file_name = "NATE" + keyword + "_list_" + time.strftime("%y%m%d_%H%M%S") + '.txt'
output_file = open(output_file_name, "w", encoding="utf-8")
output_file.write("{}\t{}\t{}\n".format('페이지', '제목', 'URL'))
output_file.close()

# 제목, URL  파일 저장
def write_news(i, article_title, article_url):
    print([i, article_title, article_url])
    output_file = open(output_file_name, "a", encoding="utf-8")
    output_file.write("{}\t{}\t{}\n".format(i, article_title, article_url))
    return

def crawl_news(i):
    page = i

    url = 'https://pann.nate.com/talk/c20048?page=' + str(page)
    # print(url)

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}
    html_req = requests.get(url, headers=headers)
    tree = html.fromstring(html_req.content)

    bodies = tree.xpath('//td[@class="subject"]')  # td class가   subject인거

    # print(len(bodies))
    print(len((bodies)))             #  게시글 개수
    results = []


    # for body in bodies:
    #     body.xpath('a/text()')[0]       # n번째 게시글의 xpath

        # print(body.xpath('a/text()'))
    for body in bodies:# 게시글 개수만큼
        try:
            article_title = body.xpath('a/text()')[0]
        except:
            article_title = ''
        # print(article_title)
        try:
            article_url = body.xpath('a/@href')[0]
            article_url = 'https://pann.nate.com'+article_url
        except:
            article_url = ''
        # print(article_url)

        if article_title != '':               #  게시글
            article_title_clean = article_title.replace("\n", "").replace("\t", "").replace("\r", "").strip()
            results.append([article_title_clean, article_url])
            write_news(i, article_title_clean, article_url)

    output_file.close()

    return results


def fmain():

    for i in range(1, 3):
        # print(i)
        results = crawl_news(i)
        # print(results)
        # time.sleep(6)


fmain()
```
